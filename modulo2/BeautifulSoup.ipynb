```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.decathlon.com/collections/deals?page=1'
```

- ¿Qué podríamos extraer de ahí?
- Botón derecho en la web e inspect o inspeccionar
- Flechita arriba a la izquierda, así cuando seleccionamos algo, se selecciona el código


```python
result = requests.get(url)

print(result.status_code)
```

```python
result.content #¡Qué horror!
```

```python
soup = BeautifulSoup(result.content, 'html.parser') #parsear es como "traducir" a HTML
```

```python
soup #Aquí está todo, pero no necesitamos todo el contenido
```

- Tenemos que volver a la web para ver qué necesitamos
- Voy al nombre del producto
- Me debe poner span Class etc.
- El código es diferente para el nombre del producto, precio, etc.
- Queremos guardar los datos en categorias y precios y crear un DataFrame con esta info

```python
lista_categorias = soup.find_all('span', {'class': 'de-u-textMedium de-u-textShrink1 de-u-lg-textGrow1 de-u-textBold'})
```

```python
lista_categorias
```

```python
categorias = []

for i in lista_categorias:
    categorias.append(i.getText()) #esto se hace hasta para los precios, sive para quitar las etiquetas
categorias 
```

- Todavía tenemos los saltos de línea y demás
- Habrá que trabajar con estas listas para limpiarla, esto ya lo sabemos hacer

```python
categorias_limpias = []
```

```python
for categoria in categorias:
    categorias_limpias.append(categoria.replace("\n", "").strip())


categorias_limpias
```

# Ahora buscamos precios
#

```python
lista_precios = soup.find_all('span', {'class': 'js-de-ProductTile-currentPrice'})
```

```python
lista_precios
```

```python
precios = []

for i in lista_precios:
    precios.append(i.getText()) 
precios
```

# Los pasos son:
- url
- llamada con request

- parte del código que necesitemos
- extraer texto
- limpiar
- listas -> diccionario -> dataframe

```python
diccionario_decathlon = {}
diccionario_decathlon['categorias'] = categorias_limpias
diccionario_decathlon['precios'] = precios
```

```python
diccionario_decathlon
```

```python
df = pd.DataFrame(diccionario_decathlon)

print(df)
```

- Extracción de datos de móviles de Ebay


```python
url_ebay = "https://www.ebay.es/e/campanas/moviles-y-smartphones-reacondicionados"
```

```python
result_ebay = requests.get(url_ebay)
result_ebay.status_code
```

```python
result_ebay.content
```

```python
soup_ebay = BeautifulSoup(result_ebay.content, 'html.parser')
```

```python
soup_ebay
```

```python
#En la web, pincho sobre el título del móvil que es la parte que quiero extraer
#h3.s-item__title
```

```python
lista_moviles = soup_ebay.find_all('h3', {'class': 'textual-display bsig__title__text'})
```

```python
lista_moviles
```

```python
#Ahora hay que extraer el texto de aquí
```

```python
moviles = []


for i in lista_moviles:
        moviles.append(i.getText())
moviles
```

- Ya tenemos la columna del nombre del producto
- Vamos con los precios

```python
lista_moviles_precios = soup_ebay.find_all('span', {'class': "textual-display bsig__price bsig__price--displayprice"})
lista_moviles_precios
```

```python
precios = []


for i in lista_moviles_precios:
        precios.append(i.getText())
precios
```

```python
precios_limpios = []
```

```python
for i in precios:
    precio = i.replace("EUR", "").replace("\xa0", "").strip().replace(",", ".")
    precio_limpio = float(precio)
    precios_limpios.append(precio_limpio)
```

```python
precios_limpios
```

```python
#Ahora hacemos
# - Un diccionario
# - Un dataframe
```

- ¿Y si quiero extraer los datos de la siguiente página?
- ¿Qué pasa al pasar la página? La URL cambia, así que habrá que cambiar la URL


```python
for pagina in range(1, 6):
 
    url_moviles = f"https://www.ebay.es/e/campanas/moviles-y-smartphones-reacondicionados?_pgn={pagina}"
    
```

```python
diccionario_ebay = {}
diccionario_ebay['nombres'] = moviles
diccionario_ebay['precios'] = precios_limpios
diccionario_ebay
```

```python
df_ebay = pd.DataFrame(diccionario_ebay)

print(df_ebay)
```

```python
# al igual que en el ejemplo anterior lo primero que haremos será definir la url de la página de la vamos a sacar datos
url_bolsa = "https://www.bolsamania.com/indice/IBEX-35/historico-precios"

# hacemos la request a la página de la que queremos sacar la info
res_bolsa = requests.get(url_bolsa)

# vemos si todo ha ido bien
res_bolsa.status_code
```

```python
res_bolsa.content
```

```python
soup_bolsa = BeautifulSoup(res_bolsa.content, 'html.parser')
```

```python
tablas = soup_bolsa.find_all("table")
tablas
```

```python
len(tablas) #a ver cuántas tablas tengo
```

```python
tablas[0]
```

```python
tablas[1]
```

```python
tablas[2]
```

```python
nuestra_tabla = tablas[2]
```

```python
#Primero vamos a conseguir los titulos de las columnas

lista_encabezados = nuestra_tabla.find_all('th') #th se refiere a encabezados, esto lo tenéis en los apuntes

lista_encabezados
```

```python
encabezados = []

for col in lista_encabezados:
    encabezados.append(col.text)

encabezados
```

```python
filas_tabla = nuestra_tabla.find_all('tr') #Aquí aparecen los encabezados y el contenido
filas_tabla
```

```python
resultados = []
for col in filas_tabla[1:]: #lo pongo así porque el título que es lo que sale primero no lo quiero
    resultados.append(col.text)
resultados
```

```python
resultados_limpios = []

for resultado in resultados:
    elemento = resultado.split("\n")
    resultados_limpios.append(elemento)
```

```python
resultados_limpios #cada resultado es una lista
```

```python
resultados_finales = []

for resultado in resultados_limpios:
    elemento = resultado[1:-1]
    resultados_finales.append(elemento)
```

```python
resultados_finales
```

```python
df_tabla = pd.DataFrame(resultados_finales)
df_tabla
```

```python
df_tabla.columns = encabezados
```
